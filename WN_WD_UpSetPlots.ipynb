```python
import pandas as pd
from upsetplot import UpSet
from upsetplot import from_memberships
import matplotlib.pyplot as plt

# Define your set memberships
memberships = [
    ('Babel',),
    ('OEWN',),
    ('GF',),
    ('Yovisto Spotlight',),
    ('Yovisto KEA',),
    ('Babel', 'OEWN'),
    ('Babel', 'GF'),
    ('Babel', 'Yovisto Spotlight'),
    ('Babel', 'Yovisto KEA'),
    ('OEWN', 'GF'),
    ('OEWN', 'Yovisto Spotlight'),
    ('OEWN', 'Yovisto KEA'),
    ('GF', 'Yovisto Spotlight'),
    ('GF', 'Yovisto KEA'),
    ('Yovisto Spotlight', 'Yovisto KEA'),
    ('Babel', 'GF', 'Yovisto Spotlight', 'Yovisto KEA'),
    ('Babel', 'OEWN', 'GF', 'Yovisto Spotlight', 'Yovisto KEA'),
]

data1 = [
    39224,
    12083,
    30418,
    36350,
    34339,
    9635,
    21782,
    26490,
    21037,
    7507,
    9593,
    7475,
    18028,
    16557,
    19826,
    9819,
    3917

]

# Define counts for each combination
data = from_memberships(memberships, data=data1)

# Create the plot
upset = UpSet(data)
upset.plot()

plt.show()
```
